```python
print(df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df['Sex']=df['Sex'].map({'male':0,'female':1})
df['Embarked']=df['Embarked'].map({'C':0,'Q':1,'S':2})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.histplot(df['Age'],kde=True)
plt.title('Distribution of Age')
plt.show()


sns.histplot(df['Fare'],kde=True)
plt.title('Distribution of Fare')
plt.show()
```

```python
plt.plot(df['Fare'],df['Age'],color='g')
plt.title('Plot')  
plt.ylabel('Fare')  
plt.xlabel('Age')  
plt.show()  
```

```python
sns.barplot(x='Sex',y='Survived',data=df)
plt.title('Survival rate by sex')
plt.show()
```

```python
sns.histplot(df[df['Survived']==1]['Age'],kde=True,label='Survived',color='r')
sns.histplot(df[df['Survived']==0]['Age'],kde=True,label='Not Survived',color='g')

plt.title('Age distribution by survival')
plt.legend()
plt.show()
```

```python
sns.barplot(x='Pclass',y='Survived',data=df)
plt.title('Survival rate by Passenger Class')
plt.show()
```

```python
sns.barplot(x='Embarked',y='Survived',data=df)
plt.title('Survival rate by Embarked Location')
plt.show()
```

```python
sns.countplot(x='Fare',data=df)
plt.title('Count by Fare')
plt.show()
```

```python
sns.boxplot(x='Sex',y='Age',data=df,palette='PuBuGn')
plt.title('Age by sex')
```

```python
sns.jointplot(x='Sex',y='Survived',data=df,kind='hex',color='Green')
```

```python
sns.distplot(df['Age'],color='Red')
plt.title('Plotting by Age')
```

```python
sns.violinplot(x='Sex',y='Survived',data=df,split=True)
plt.title('Survival by Sex')
```

```python
corr_matrix=df.corr()

sns.heatmap(corr_matrix,annot=True,cmap='Oranges')
plt.title('Correlation Matrix')
plt.show()
```
